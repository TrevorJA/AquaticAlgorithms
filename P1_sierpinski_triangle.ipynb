### Step 3: Coding!

The first step is to import the necessary Python libraries that we will use in this program. We will use numpy for numerical calculations, matplotlib.pyplot for plotting, and random for generating random numbers.

```python
# Core modules
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
```

Okay, now we need to start by defining the three corners.   I use some basic (x,y) coordinates to set the corners and call them (`A`, `B`, and `C`). 

```python
# Corners (x, y) coordinates
A = [0, 0]
B = [1, 0]
C = [0.5, np.sqrt(0.75)]

# Put them all in a list
list_of_corners = [A, B, C]
list_of_corners
```

In the end, we want two lists which contain all of the coordinates for points.  We will put all the x-coordinates in list, and all the y-coordinates in another.  

Below, I start the list (a blank set of `[]`) and then I use the `[].append()` function to add each coordinate point to the list.  The `append()` function will simply add a number to the end of the list.  

Let's start by adding the coordinates for our corner points:

```python
## Store the x and y coordinates for the corners inside our list

# Make the empty lists
x_coordinates= []
y_coordinates= []

# Add cornrer coordinates to a list

# Loop 3 times, one for each corner
for i in [0,1,2]:
    # Append the values to the lists
    x_coordinates.append(list_of_corners[i][0])
    y_coordinates.append(list_of_corners[i][1])

# list_of_corners[1][0]
```

Great, we are off to a good start!  

The next section is where the challenge is!    

We are going to repeat the steps shown in the video to make N random points on the triangle.  Any time we need to repeat something a bunch of times, you should think: "**for-loop**".     

Below, I give the outline of the algorithm, the same way that it was set up in the video... see if you can replace the `None` values with the coded needed to make it all work!

```python
### Generate data for the sierpinski triangle

# How many points do you want?
n_points = 1000000

# Set the first point inside the triangle
new_point = [0.5, 0.5] 

# Start the for loop and repeat the processs N times
for i in range(n_points):

    ## Choose a random corner
    # `random.randint` returns a random number between the two numbers you give it
    corner_number = random.randint(0,2)
    
    # Get the corner coordinate points from the list
    # Think, we just chose a random number, now we need to use that number to get the specific corner from the list
    corner_points = list_of_corners[corner_number]

    ## Calculate the half-way point between the current point and the corner
    half_way_point = [(new_point[0]+corner_points[0])/2,
                      (new_point[1]+corner_points[1])/2]


    ## Add the half way point to the data
    x_coordinates.append(half_way_point[0])
    y_coordinates.append(half_way_point[1])

    ## Set the half way point as the new starting point
    new_point = half_way_point
    
    # Great, start over from the top of the for-loop!

print(f'Great, the for-loop is done and you have {len(x_coordinates)} data points! Lets plot!')
```

```python
# Initialize figure
fig, ax = plt.subplots(figsize = (7,7), dpi=150)

# Set style
ax.set(xlim = (-0.1, 1.1), ylim = (-0.1, 1))
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)

# Background color
ax.set_facecolor('tan')

# Set the title
ax.set_title(f'The Sierpinski Triangle\n{n_points} Points', fontsize=16)

# Add the dots
ax.scatter(x_coordinates, y_coordinates, 
           color = 'black', 
           marker='.',
           s = 0.1)

plt.show()

plt.savefig(f'Sierpinski_triangle_{n_points}_points.png', dpi = 200)
```
